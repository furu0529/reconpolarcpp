```python
def make_array(grid,L,arraymap):
    real=np.zeros((grid,grid,grid))
    imag=np.zeros((grid,grid,grid))
    for i in range(grid):
        for j in range(grid):
            for k in range(grid):
                real[i,j,k]=arraymap[i*grid*grid+j*grid+k][0]
                imag[i,j,k]=arraymap[i*grid*grid+j*grid+k][1]
    return real,imag

def make_karray(grid,L,arraymap):
    bins=int (grid/2)
    real=np.zeros((grid,grid,bins))
    #imag=np.zeros((grid,grid,bins))
    for i in range(grid):
        for j in range(grid):
            for k in range(bins):
                real[i,j,k]=arraymap[i*grid*bins+j*bins+k]
                #imag[i,j,k]=arraymap[i*grid*grid+j*grid+k-1][1]
    return real#,imag
```
